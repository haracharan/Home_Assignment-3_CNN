```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.optimizers import Adam

# Load and normalize MNIST
(x_train, _), (x_test, _) = mnist.load_data()
x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.
x_train = x_train.reshape((len(x_train), 784))
x_test = x_test.reshape((len(x_test), 784))

# Function to build autoencoder
def build_autoencoder(latent_dim):
    input_img = Input(shape=(784,))
    encoded = Dense(latent_dim, activation='relu')(input_img)
    decoded = Dense(784, activation='sigmoid')(encoded)
    autoencoder = Model(input_img, decoded)
    return autoencoder

# Train and visualize for different latent dimensions
latent_dims = [16, 32, 64]
for latent_dim in latent_dims:
    print(f"\nTraining Autoencoder with latent dimension: {latent_dim}")
    autoencoder = build_autoencoder(latent_dim)
    autoencoder.compile(optimizer=Adam(), loss='binary_crossentropy')
    autoencoder.fit(x_train, x_train,
                    epochs=10,
                    batch_size=256,
                    shuffle=True,
                    validation_data=(x_test, x_test),
                    verbose=0)

    # Predict on test data
    reconstructed = autoencoder.predict(x_test)

    # Plot original vs reconstructed
    n = 10
    plt.figure(figsize=(20, 4))
    plt.suptitle(f"Latent Dimension: {latent_dim}", fontsize=16)
    for i in range(n):
        # Original
        ax = plt.subplot(2, n, i + 1)
        plt.imshow(x_test[i].reshape(28, 28), cmap='gray')
        plt.title("Original")
        plt.axis('off')

        # Reconstructed
        ax = plt.subplot(2, n, i + 1 + n)
        plt.imshow(reconstructed[i].reshape(28, 28), cmap='gray')
        plt.title("Reconstructed")
        plt.axis('off')
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.optimizers import Adam

# Load and normalize MNIST
(x_train, _), (x_test, _) = mnist.load_data()
x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.
x_train = x_train.reshape((len(x_train), 784))
x_test = x_test.reshape((len(x_test), 784))

# Add Gaussian noise (mean=0, std=0.5) to inputs
noise_factor = 0.5
x_train_noisy = x_train + np.random.normal(loc=0.0, scale=noise_factor, size=x_train.shape)
x_test_noisy = x_test + np.random.normal(loc=0.0, scale=noise_factor, size=x_test.shape)

# Clip to maintain valid pixel range [0, 1]
x_train_noisy = np.clip(x_train_noisy, 0., 1.)
x_test_noisy = np.clip(x_test_noisy, 0., 1.)

# Function to build autoencoder
def build_autoencoder(latent_dim=32):
    input_img = Input(shape=(784,))
    encoded = Dense(latent_dim, activation='relu')(input_img)
    decoded = Dense(784, activation='sigmoid')(encoded)
    return Model(input_img, decoded)

# === 1. Train Basic Autoencoder ===
print("\nTraining Basic Autoencoder...")
basic_autoencoder = build_autoencoder()
basic_autoencoder.compile(optimizer=Adam(), loss='binary_crossentropy')
basic_autoencoder.fit(x_train, x_train,
                      epochs=10,
                      batch_size=256,
                      shuffle=True,
                      validation_data=(x_test, x_test),
                      verbose=0)
basic_reconstructed = basic_autoencoder.predict(x_test)

# === 2. Train Denoising Autoencoder ===
print("\nTraining Denoising Autoencoder...")
denoising_autoencoder = build_autoencoder()
denoising_autoencoder.compile(optimizer=Adam(), loss='binary_crossentropy')
denoising_autoencoder.fit(x_train_noisy, x_train,
                          epochs=10,
                          batch_size=256,
                          shuffle=True,
                          validation_data=(x_test_noisy, x_test),
                          verbose=0)
denoised_reconstructed = denoising_autoencoder.predict(x_test_noisy)

# === 3. Visualization: Noisy vs Denoised vs Original ===
n = 10
plt.figure(figsize=(20, 6))
plt.suptitle("Denoising Autoencoder Results", fontsize=16)

for i in range(n):
    # Noisy input
    ax = plt.subplot(3, n, i + 1)
    plt.imshow(x_test_noisy[i].reshape(28, 28), cmap='gray')
    plt.title("Noisy")
    plt.axis('off')

    # Denoised output
    ax = plt.subplot(3, n, i + 1 + n)
    plt.imshow(denoised_reconstructed[i].reshape(28, 28), cmap='gray')
    plt.title("Denoised")
    plt.axis('off')

    # Original clean
    ax = plt.subplot(3, n, i + 1 + 2 * n)
    plt.imshow(x_test[i].reshape(28, 28), cmap='gray')
    plt.title("Original")
    plt.axis('off')
plt.tight_layout()
plt.show()

# === 4. Visualization: Clean Input vs Basic Output ===
plt.figure(figsize=(20, 4))
plt.suptitle("Basic Autoencoder Results", fontsize=16)
for i in range(n):
    ax = plt.subplot(2, n, i + 1)
    plt.imshow(x_test[i].reshape(28, 28), cmap='gray')
    plt.title("Input")
    plt.axis('off')

    ax = plt.subplot(2, n, i + 1 + n)
    plt.imshow(basic_reconstructed[i].reshape(28, 28), cmap='gray')
    plt.title("Reconstructed")
    plt.axis('off')
plt.tight_layout()
plt.show()

# === 5. Real-world Use Case Explanation ===
print("\nUse Case: Denoising autoencoders are valuable in real-world scenarios like medical imaging. For example, when MRI or CT scans are corrupted with noise due to low radiation dosage or sensor interference, denoising autoencoders can help reconstruct clean, usable images without losing critical diagnostic details.")
```
